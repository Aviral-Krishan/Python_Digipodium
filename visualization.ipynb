visualization
- line plot(time series, trend, seasonality, cyclicity, noise, outliers, missing valuses etc)
- area plot(similar to line plot, but area under the line is emphasized)
- bar plot (categorical data, frequency, distribution etc for continous data comparison)
- histogram(contimous data, frequency, frequency distribution etc for continous daa comparison)
- pie chart(same as bar chart but smaller number of categories and percentage comparison)
- box plot(statistical evaluation outliers,outliers, quartiles etc for continous data comparison)
- scatter plot(correlation, relationship, etc for continous data comparison)
- bubble plot(scatter plot with additional dimention of data)
